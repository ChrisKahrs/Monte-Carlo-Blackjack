```python
import pandas as pd
import matplotlib.pyplot as plt
import random
```

```python
# Find normal distribution when rolling dice twice and summing results.

results = {}

# Run many, many times to exhaust all possibilities
for i in range(100000):
    roll_1 = random.randint(1,6)
    roll_2 = random.randint(1,6)
    
    sum_dice = roll_1 + roll_2
    
    results[sum_dice] = results.get(sum_dice, 0) + 1
    

# Load frequency of sums to a df
df = pd.DataFrame(list(results.values()), columns = ["sums"], index = results.keys())
# sort by index to get a smooth graph
df.sort_index(inplace=True)

```

```python
df.plot.bar()

plt.xlabel('Dice Sum')
plt.ylabel('Frequency')

plt.show()
```
